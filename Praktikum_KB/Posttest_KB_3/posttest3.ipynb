```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
data = pd.read_csv('heart.csv')
data.info()
```

```python
data.head()
```

visualiasi data menggunakan
# grafik pie

```python
data = data.dropna(subset=['age'])
```

```python
data_heart = data['age'].value_counts().rename_axis('age').reset_index(name='Id_count')
```

```python
plt.figure(figsize=(10,10))
plt.pie(
    data_heart.Id_count, 
    labels     = data_heart.age, 
    startangle = 10,
    autopct    = '%.1f%%',
    shadow=True
)
plt.title('Data yang mempunyai penyakit hati berdasarkan Kategori Usia')
plt.show()
```

visualisasi data menggunakan
# heatmap

```python
cor_matrix = data_heart.corr()
sns.heatmap(cor_matrix, cmap='coolwarm', annot=True,  center=0)
plt.show
```

visualisasi data menggunakan
# grafik plot


```python
plt.figure(figsize=(5,10))
sns.boxplot(
    data   = data,
    x      = 'exang',
    y      = 'age',
)

plt.xlabel('exang')
plt.ylabel('age')
plt.title('Orang yang Memiliki sakit hati atau exang berdasarkan umur')
plt.show()
```

dari grafik plot diatas menunjukkan umur antara 40an hingga 60an rata rata mempunyai penyakit hati
dan 50an hingga 60an yang paling banyak terkena penyakit tersebut

visualisasi data menggunakan
# scatter plot

```python
dataplotting=data.sort_values("exang").head(50)
plt.xticks(rotation='horizontal', size=10)
sns.scatterplot(x="age",y="exang",data=dataplotting) 
```

dari hasil visualisasi diatas menggunakan scatter plot dapat dilihat bahwa kebanyakan yang terknea penyakit hati atau exang berada rentan pada umur 40an hingga 60an
