### Import Library

```python
import pandas as pd
import numpy as np 
from sklearn.model_selection import train_test_split
```

### menangani nilai 0

```python
from sklearn.impute import SimpleImputer
```

```python
imputer = SimpleImputer(strategy='mean')
dataset_split["age"] = imputer.fit_transform(dataset_split[["age"]])
```

```python
dataset_split.head()
```

```python
print("jumlah null setelah menggunakan simpleimputer")
print(dataset_split.isna().sum())
```

```python
dataset_split.shape
```

## membagi training set dan testing set dengan ketentuan 70:30

```python
X = dataset_split.iloc[:,:-1]
y = dataset_split.iloc[:,-1]

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = 0.3)
```

```python
print("Dataset dibagi dengan proporsi 70:30\n")
print("Dimensi X_train: ", X_train.shape)
print("Dimensi X_test: ", X_test.shape)
print("Dimensi y_train: ", y_train.shape)
print("Dimensi y_test: ", y_test.shape)
```

### MELAKUKAN DATA CLEANING PADA DATA DENGAN NILAI NULL

```python
print ("Cek Tipe Data :")
dataset_split_rusak.dtypes
```

### MENGGANTI NILAI NULL SESUAI KETENTUAN

```python
from sklearn.impute import SimpleImputer
```

```python
imputer = SimpleImputer(strategy = "mean")

dataset_split_rusak['age'] = imputer.fit_transform(dataset_split_rusak[['age']])

dataset_split_rusak.isna().sum()
```

```python
imputer = SimpleImputer(strategy = "most_frequent")

dataset_split_rusak['sex'] = imputer.fit_transform(dataset_split_rusak[['sex']])

dataset_split_rusak.isna().sum()
```

```python
imputer = SimpleImputer(strategy = "mean")

dataset_split_rusak['cp'] = imputer.fit_transform(dataset_split_rusak[['cp']])

dataset_split_rusak.isna().sum()
```

```python
imputer = SimpleImputer(strategy = "median")

dataset_split_rusak['exang'] = imputer.fit_transform(dataset_split_rusak[['exang']])

dataset_split_rusak.isna().sum()
```
